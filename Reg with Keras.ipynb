```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
data.head()
```

```python
X=data.drop(columns=["Strength"])
y=data["Strength"]
```

```python
X.head()
```

```python
y.head()
```

```python
X_norm=(X-X.mean())/X.std()

X_norm.head()
```

```python
model=Sequential()
model.add(Dense(5,activation="relu",input_shape=(X.shape[1],)))
model.add(Dense(5,activation="relu"))
model.add(Dense(1))
```

```python
model.compile(optimizer="adam",loss="mean_squared_error")
```

```python
model.fit(X,y,validation_split=0.2,epochs=150)
```
